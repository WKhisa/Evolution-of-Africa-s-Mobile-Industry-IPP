```python
# Import pandas for data manipulation
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# Get data
raw_df= pd.read_csv('https://bit.ly/MobileDataset')
```

```python
#Preview data
raw_df.info()
```

```python
raw_df.head()
```

```python
#clean data
clean_df = raw_df.replace('..',np.nan)
clean_df.head()
```

```python
columns_float = ['1995 [YR1995]', '1996 [YR1996]', '1997 [YR1997]', '1998 [YR1998]','1999 [YR1999]', '2000 [YR2000]', '2001 [YR2001]', '2002 [YR2002]','2003 [YR2003]', '2004 [YR2004]', '2005 [YR2005]', '2006 [YR2006]','2007 [YR2007]', '2008 [YR2008]', '2009 [YR2009]', '2010 [YR2010]','2011 [YR2011]', '2012 [YR2012]', '2013 [YR2013]', '2014 [YR2014]','2015 [YR2015]', '2016 [YR2016]']
clean_df[columns_float]= clean_df[columns_float].astype(float)
clean_df.info()
```

```python
clean_df.drop(columns= ['1995 [YR1995]', '1996 [YR1996]', '1997 [YR1997]', '1998 [YR1998]','1999 [YR1999]','2016 [YR2016]'], inplace= True)
clean_df.info()
```

```python
#Trends in Internet Penetration
internet_penetration_df = clean_df[clean_df['Series Name']=='Internet users (per 100 people)'].describe().T.reset_index()
internet_penetration_df
```

```python
fig = px.bar(internet_penetration_df, x=internet_penetration_df['index'], y=internet_penetration_df['mean'], title='Trends in Internet Penetration')
fig.update_xaxes(title_text='Year')
fig.update_yaxes(title_text='Mean Value of Internet users per 100 persons')
fig.show()
```

```python
#Mobile vs Fixed Broadband subscriptions
clean_df[(clean_df['Series Name'] == 'Mobile cellular subscriptions') | (clean_df['Series Name'] == 'Fixed broadband subscriptions')]
```

```python
mobile_subs_df = clean_df[(clean_df['Series Name'] == 'Mobile cellular subscriptions')].describe().T.reset_index()
fixed_subs_df = clean_df[(clean_df['Series Name'] == 'Fixed broadband subscriptions')].describe().T.reset_index()
```

```python
# Create line charts using Plotly Express
fig = px.line(mobile_subs_df, x=mobile_subs_df['index'], y=mobile_subs_df['mean'], title='Fixed vs Mobile Subscriptions')
fig.add_scatter(x=fixed_subs_df['index'], y=fixed_subs_df['mean'], mode='lines', name='Fixed Broadband')
# Customize the axis labels
fig.update_xaxes(title_text='Year')
fig.update_yaxes(title_text='Subscriptions')

# Show the plot
fig.show()
```

```python
ict_exports_df = clean_df[(clean_df['Series Name'] == 'ICT goods exports (% of total goods exports)')].describe().T.reset_index()
ict_imports_df = clean_df[(clean_df['Series Name'] == 'ICT goods imports (% total goods imports)')].describe().T.reset_index()
fig = px.line(ict_exports_df, x=ict_exports_df['index'], y=ict_exports_df['mean'], title='ICT Imports vs Exports')
fig.add_scatter(x=ict_imports_df['index'], y=ict_imports_df['mean'], mode='lines', name='Imports')
# Customize the axis labels
fig.update_xaxes(title_text='Year')
fig.update_yaxes(title_text='Subscriptions')

# Show the plot
fig.show()
```
